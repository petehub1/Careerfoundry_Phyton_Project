# TASK 3 You need to provide the Instacart senior stakeholders with descriptive findings about sales. Create a histogram of the “order_hour_of_day” column.
In a markdown cell beneath your histogram, describe what the histogram demonstrates.

```python
hist = ords_prods_customers_merge['order_hour_of_day'].plot.hist(bins = 24)
```

Given that a day consists of 24 hours, I divided the data into 24 bins. The resulting histogram illustrates the frequency of orders for each hour of the day. We observe a lower frequency of orders during the night, indicating reduced activity. As the morning progresses, the number of orders begins to rise, peaking between the late morning and early afternoon (approximately 11 AM to 5 PM). Following this peak, there is a gradual decline in order frequency through the late afternoon and evening hours.

In conclusion, the data suggests that customers predominantly prefer to use Instacart during the late morning and early afternoon hours.
